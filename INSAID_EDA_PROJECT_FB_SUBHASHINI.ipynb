## Reading the face book dataset

```python
fb_data=pd.read_csv("https://raw.githubusercontent.com/insaid2018/Term-1/master/Data/Projects/facebook_data.csv")
fb_data.head()
```

## Metadata explanation of the columns

- userid                    -- User id generated by facebook for each user, which should be unique<br/>
- age                       -- Age of the person while joining the face book   
- dob_day                   -- Day of birth (Ranges from 1- 31)   
- dob_year                  -- Year of birth (In this dataset it is ranging from 1900 - 2000)   
- dob_month                 -- Month of birth (Ranges from Jan - Dec)   
- gender                    -- Gender of the person who joined Facebook (Male/Female)   
- tenure                    -- For how many no. of days a person use the  Facebook or active on Facebook   
- friend_count              -- Friend count of each person in the Facebook   
- friendships_initiated     -- Friendships initiated by each person in the Facebook   
- likes                     -- Total Likes given (Sum of mobile_likes and www_likes)   
- likes_received            -- Total Likes received  (sum of mobile_likes_received and www_likes_received)   
- mobile_likes              -- Likes given thru mobile    
- mobile_likes_received     -- Likes received thru mobile   
- www_likes                 -- Likes given thru the website    
- www_likes_received        -- Likes received thru the website

## Pre Profiling


```python

profile = pandas_profiling.ProfileReport(fb_data)
profile.to_file(output_file="facebook_EDA_before_preprocessing.html")
```

**Below are the pre profiling observations**

**Numeric Columns** - userid,age,dob_day,dob_month,dob_year,friend_count,friendships_initiated,likes,
likes_received,mobile_likes,mobile_likes_received,tenure,userid,www_likes,www_likes_received    

**Categorical Columns** - Gender

**Note** : Since there are not much insights can be drawn on userid,dob_year columns, we can ignore these columns for now.

**Observations**
    
- Gender has Male,Female and NA values( NA values either can be changed to Male based on mode or can ignore 175 NA records as they are only 0.17%  out of total records).    
- Tenure is having 2 null values which can be ignored which is again a minimal records when compared to total no. of records in the dataset.                                          
- likes_received is highly skewed (As per profiling report).    
- mobile_likes_received is highly correlated with likes_received which is 0.97 (As per profiling report).    
- www_likes_received is highly correlated with likes_received which is 0.94 (As per profiling report).    
- Though friend_count,friendships_initiated,likes,likes_received,mobile_likes,mobile_likes_received,www_likes, www_likes_received has some zeros. With this data we can infer if there can be any chance of having dummy accounts.    



```python
# Dropping userid  column as it unique column and no insights can be drawn from it
# Dropping dob_month column as we can't draw any insights from these.
# Removing 2 records based on tenure column which are null
# Removing 175 rows whose gender is null
# Creating new column age_group based on frequency
# Tenure is available in no. of days, convert them into no. of years

fb_data.drop(['userid','dob_year'],axis=1,inplace=True)
fb_data=fb_data[~(fb_data['tenure'].isnull())]
fb_data=fb_data[~(fb_data['gender'].isnull())]
labels=['13-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','101-110','111-120']
fb_data['age_group'] = pd.cut(fb_data.age,bins=np.arange(10,121,10),labels=labels,right=True).sort_values()
labels1=['0-1','1.1-2','2.1-3','3.1-4','4.1-5','5.1-6','6.1-7','7.1-8','8.1-9']
fb_data['tenure_group'] = pd.cut(fb_data.tenure/365,bins=np.arange(0,10,1),labels=labels1,right=True).sort_values()



```

## Observation 7 . Which is used more either desktop or Mobile to access Facebook

```python
# Observation 7 How many are using desktop and app for fb based on gender?
labels="Mobile Usage","Desktop Usage"
explode = (0, 0.1,)
fb_data["mobile"]=pd.Series(np.where(fb_data["mobile_likes"]>0,1,0)).astype('category')
fb_data.mobile.value_counts()
frac=(fb_data.mobile==1).sum()/fb_data.mobile.size
print('No. of users using Facebook thru Mobile app is {:12.2f}'.format(frac))
fb_data["desktop"]=pd.Series(np.where(fb_data["www_likes"]>0,1,0)).astype('category')
fb_data.desktop.value_counts()
frac1=(fb_data.desktop==1).sum()/fb_data.desktop.size
print('No. of users using Facebook thru desktop app is {:12.2f}'.format(frac1))
series=np.array([frac,frac1])
plt.pie(series,autopct='%1.1f%%',pctdistance=0.6,labels=labels,explode=explode)
plt.title("Chart showing usage of Mobile app and Desktop",fontweight='bold',fontsize=15)
```

#### More than 60% of users are using Mobile than desktop to access Facebook

####  Access of Facebook thru mobile or desktop based on gender

### Observation 8. Detecting Fake/Dummy accounts

```python
# Detecting Fake/Dummy Accounts
fb_data.loc[(fb_data.friend_count==0) & (fb_data.tenure==0)]['dob_day'].value_counts().plot(kind='line',color='blue',figsize=(15,7))
fb_data.loc[(fb_data.friend_count==0) & (fb_data.tenure==0)]['dob_month'].value_counts().plot(kind='line',color='green',figsize=(15,7))
plt.xlabel('dob_day/dob_month')
plt.ylabel('No. of user on day/month wise')
plt.title('Detecting Fake accounts using friend count & tenure ',fontsize=15,fontweight='bold')
fb_data["mobile"]=pd.Series(np.where(fb_data["mobile_likes"]>0,1,0)).astype('category')
fb_data.mobile.value_counts()
frac=(fb_data.mobile==1).sum()/fb_data.mobile.size
```

#### Blue line represents day_day , 14 users whose friendcount is 0 are having dob_day as first day of month as their birth date and Green line represents day_month,16 users whose friendcount is 0 are having dob_month as first month of the year and can make a insight like either these users are having dummy/Fake accounts in Facebook

```python
a=fb_data.loc[(fb_data.friend_count==0) & (fb_data.tenure==0)]['dob_day'].value_counts()
a=a.iloc[0].astype(int)
b=fb_data.loc[(fb_data.friend_count==0) & (fb_data.tenure==0)]['dob_day'].value_counts().sum()
b=b.astype(int)
c=a/b
print('No. of users whose date is in date of birth is {:12.2f}'.format(c))
d=fb_data.loc[(fb_data.friend_count==0) & (fb_data.tenure==0)]['dob_month'].value_counts()
d=d.iloc[0].astype(int)
e=fb_data.loc[(fb_data.friend_count==0) & (fb_data.tenure==0)]['dob_month'].value_counts().sum()
e=b.astype(int)
f=d/e
g=1-(c+f)
print('No. of users whose month in date of birth is {:12.2f}'.format(f))
labels="First day of Month","First Month of year","Rest"
explode = (0.1, 0.1,0.1)
series=np.array([c,f,g])
plt.pie(series,autopct='%1.1f%%',pctdistance=0.6,labels=labels,explode=explode)
plt.title("Chart showing usage of Mobile app and Desktop",fontweight='bold',fontsize=15)
```

#### In total 100% of users whose friend coutn is 0,  there are 50% of total users  whose friend count is 0 where  their  dob_day falling on 1st of month and and dob_month falling on 1st month of the year indicates either these accounts or dummy or fake accounts
